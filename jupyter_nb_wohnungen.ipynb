# Feature Transformation

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from xgboost import XGBRegressor
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Convert 'Einzug' to datetime format (assuming 'dd/mm/yy')
df_bereinigt['Einzug'] = pd.to_datetime(df_bereinigt['Einzug'], format='%d/%m/%y')

# Calculate the reference date (earliest date in the dataset)
reference_date = df_bereinigt['Einzug'].min()

# Calculate 'Einzug' as the duration in months since the reference date
df_bereinigt['Einzug'] = ((df_bereinigt['Einzug'] - reference_date) / pd.Timedelta(days=1)).astype(int) // 30

# Print the updated DataFrame to verify
print(df_bereinigt)
```

```python
features = [
    'Wohngebiet', 'Geschoss', 'Wohnfläche', 'Raumanzahl', 'Dusche', 'Wanne', 
    'Aufzug', 'Balkon/Terasse', 'Einstufung Mietkosten', 'Geschlecht', 
    'Generation', 'Altereinstufung_Einzug', 'Mieteralter_Einzug', 'Personenanzahl', 
    'Haustier', 'Mahnverfahren'
]
target = 'Wohndauer in Monaten'

X = df_bereinigt[features]
y = df_bereinigt[target]
```

Deskriptive Statistiken

```python
# Assuming your DataFrame is named df_bereinigt
df_features = df_bereinigt[features].describe()

print(df_features)
```

```python
numerical_features = ['Wohnfläche', 'Raumanzahl', 'Mieteralter_Einzug', 'Personenanzahl']
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_features = [
    'Wohngebiet', 'Geschoss', 'Dusche', 'Wanne', 'Aufzug', 
    'Balkon/Terasse', 'Einstufung Mietkosten', 'Geschlecht', 'Generation', 
    'Altereinstufung_Einzug', 'Haustier', 'Mahnverfahren'
]
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
# Select numerical and categorical features along with the target
numerical_data = df_bereinigt[numerical_features + [target]]
categorical_data = df_bereinigt[categorical_features + [target]]

# Compute the correlation matrix for numerical features
numerical_corr = numerical_data.corr()[target].drop(target).sort_values(ascending=False)

# One-hot encode categorical features
categorical_data_encoded = pd.get_dummies(categorical_data, columns=categorical_features)

# Compute the correlation matrix for categorical features
categorical_corr = categorical_data_encoded.corr()[target].drop(target).sort_values(ascending=False)

# Print the highest correlations for numerical features
print("Top correlations with 'Wohndauer in Monaten' for numerical features:")
print(numerical_corr.head(10))

# Print the highest correlations for categorical features
print("\nTop correlations with 'Wohndauer in Monaten' for categorical features:")
print(categorical_corr.head(10))
```

```python
# Assuming preprocessor and X, y are defined as before

# Split the data into training and test sets (60% train, 40% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Define the Ridge regression model pipeline
ridge_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', Ridge(alpha=1.0))  # You can adjust the alpha parameter
])

# Perform cross-validation on the training set
ridge_cv_scores = cross_val_score(ridge_model, X_train, y_train, cv=5, scoring='neg_mean_absolute_error')

# Print cross-validation results
print(f'Cross-Validation Mean MAE (on training set): {-ridge_cv_scores.mean()}')

# Fit the model on the training data
ridge_model.fit(X_train, y_train)

# Predictions on the test dataset
y_pred_ridge = ridge_model.predict(X_test)

# Calculate residuals on the test set
residuals_ridge = y_test - y_pred_ridge

# Evaluate the model performance on the test set
mae_ridge = mean_absolute_error(y_test, y_pred_ridge)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

print("Ridge Regression Model:")
print(f'Mean Absolute Error (on test set): {mae_ridge}')
print(f'Mean Squared Error (on test set): {mse_ridge}')
print(f'R^2 Score (on test set): {r2_ridge}')
```

```python

# Define the Lasso regression model pipeline
lasso_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', Lasso(alpha=0.1))  # You can adjust the alpha parameter
])

# Perform cross-validation on the training set
lasso_cv_scores = cross_val_score(lasso_model, X_train, y_train, cv=5, scoring='neg_mean_absolute_error')

# Print cross-validation results
print(f'Cross-Validation Mean MAE (on training set): {-lasso_cv_scores.mean()}')

# Fit the model on the training data
lasso_model.fit(X_train, y_train)

# Predictions on the test dataset
y_pred_lasso = lasso_model.predict(X_test)

# Calculate residuals on the test set
residuals_lasso = y_test - y_pred_lasso

# Evaluate the model performance on the test set
mae_lasso = mean_absolute_error(y_test, y_pred_lasso)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

print("Lasso Regression Model:")
print(f'Mean Absolute Error (on test set): {mae_lasso}')
print(f'Mean Squared Error (on test set): {mse_lasso}')
print(f'R^2 Score (on test set): {r2_lasso}')
```

```python

# Define the Linear regression model pipeline
linear_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

# Perform cross-validation on the training set
linear_cv_scores = cross_val_score(linear_model, X_train, y_train, cv=5, scoring='neg_mean_absolute_error')

# Print cross-validation results
print(f'Cross-Validation Mean MAE (on training set): {-linear_cv_scores.mean()}')

# Fit the model on the training data
linear_model.fit(X_train, y_train)

# Predictions on the test dataset
y_pred_linear = linear_model.predict(X_test)

# Calculate residuals on the test set
residuals_linear = y_test - y_pred_linear

# Evaluate the model performance on the test set
mae_linear = mean_absolute_error(y_test, y_pred_linear)
mse_linear = mean_squared_error(y_test, y_pred_linear)
r2_linear = r2_score(y_test, y_pred_linear)

print("Linear Regression Model:")
print(f'Mean Absolute Error (on test set): {mae_linear}')
print(f'Mean Squared Error (on test set): {mse_linear}')
print(f'R^2 Score (on test set): {r2_linear}')
```

```python
# Plotting the residual plot for linear regression
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_pred_linear, y=residuals_linear)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Linear Regression - Residuenplot')
plt.xlabel('Vorausgesagte Werte')
plt.ylabel('Residuen')
plt.show()
```

```python
# Define the RandomForestRegressor model pipeline
rf_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(n_estimators=100, random_state=42))
])

# Perform cross-validation on the training set
rf_cv_scores = cross_val_score(rf_model, X_train, y_train, cv=5, scoring='neg_mean_absolute_error')

# Print cross-validation results
print(f'Cross-Validation Mean MAE (on training set): {-rf_cv_scores.mean()}')

# Fit the model on the entire dataset
rf_model.fit(X, y)

# Predictions on the entire dataset (you may want to use X_test for testing, but this is how it is in your example)
y_pred_rf = rf_model.predict(X)

residuals_linear = y_test - y_pred_linear

# Evaluate the model performance
mae_rf = mean_absolute_error(y, y_pred_rf)
mse_rf = mean_squared_error(y, y_pred_rf)
r2_rf = r2_score(y, y_pred_rf)

print("RandomForestRegressor Model:")
print(f'Mean Absolute Error (on full dataset): {mae_rf}')
print(f'Mean Squared Error (on full dataset): {mse_rf}')
print(f'R^2 Score (on full dataset): {r2_rf}')
```

```python
# Calculate residuals
residuals_rf = y - y_pred_rf

# Plotting the residual plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_pred_rf, y=residuals_rf)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('RandomForestRegressor Residual Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()
```

```python
# Get feature importances
feature_importances = rf_model.named_steps['regressor'].feature_importances_

# Get column names after one-hot encoding
categorical_transformer = rf_model.named_steps['preprocessor'].named_transformers_['cat']
encoded_categorical_columns = list(categorical_transformer.named_steps['onehot'].get_feature_names_out(categorical_features))

# Combine numerical and categorical feature names
all_feature_names = numerical_features + encoded_categorical_columns

# Sort feature importances in descending order
sorted_indices = np.argsort(feature_importances)[::-1]
sorted_feature_importances = feature_importances[sorted_indices]
sorted_feature_names = np.array(all_feature_names)[sorted_indices]

# Plotting the feature importances
plt.figure(figsize=(10, 8))
plt.barh(range(len(sorted_feature_importances)), sorted_feature_importances, align='center')
plt.yticks(range(len(sorted_feature_importances)), sorted_feature_names)
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Random Forest Regressor - Feature Importances')
plt.gca().invert_yaxis()  # Invert y-axis to have the most important feature on top
plt.show()
```

```python

# Define the XGBoost model pipeline
xgboost_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', XGBRegressor(n_estimators=100, random_state=42))
])

# Perform cross-validation on the training set
xgboost_cv_scores = cross_val_score(xgboost_model, X_train, y_train, cv=5, scoring='neg_mean_absolute_error')

# Print cross-validation results
print(f'Cross-Validation Mean MAE (on training set): {-xgboost_cv_scores.mean()}')

# Fit the model on the entire dataset
xgboost_model.fit(X, y)

# Predictions on the entire dataset (you may want to use X_test for testing, but this is how it is in your example)
y_pred_xgboost = xgboost_model.predict(X)

# Evaluate the model performance
mae_xgboost = mean_absolute_error(y, y_pred_xgboost)
mse_xgboost = mean_squared_error(y, y_pred_xgboost)
r2_xgboost = r2_score(y, y_pred_xgboost)

print("XGBoost Regression Model:")
print(f'Mean Absolute Error (on full dataset): {mae_xgboost}')
print(f'Mean Squared Error (on full dataset): {mse_xgboost}')
print(f'R^2 Score (on full dataset): {r2_xgboost}')
```
